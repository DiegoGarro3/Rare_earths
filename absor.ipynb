```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_excel("data/absorbancia.xlsx", skiprows=1)
df.dropna(inplace=True)
```

```python
df.tail()
```

```python
df.describe()
```

### Vamos a eliminar las columnas repetidas que son los valores de la longitud de onda, sólo dejaremos una. 

```python
df = df.drop(df.columns[[2,4,6,8, 10]], axis=1)
```

```python
df
```

```python
fig, ax=plt.subplots(1,1, figsize=(20,15))
ax.plot(df.iloc[:,0], df.iloc[:,1], lw=3, label="Germanato")
ax.plot(df.iloc[:,0], df.iloc[:,2], lw=3, label='Matriz Pr')
ax.plot(df.iloc[:,0], df.iloc[:,3], lw=3, label='Matriz TiO2')
ax.plot(df.iloc[:,0], df.iloc[:,4], lw=3, label='Matriz ')
ax.plot(df.iloc[:,0], df.iloc[:,5], lw=3)
ax.plot(df.iloc[:,0], df.iloc[:,6], lw=3)
ax.set_xlabel("Longitud de onda [nm]", fontsize=15)
ax.set_ylabel("Absorbancia [u.a.]", fontsize=15)
ax.set_title("Absorbancia de las muestras", fontsize=25)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)
ax.grid()
ax.legend(fontsize=15)
plt.show()
```
